```python
%matplotlib ipympl

import glob
import matplotlib.pyplot as plt

trace_files = "../../path-to-trace-directory/trace_*.csv"

trace_files = sorted(glob.glob(trace_files))

traces = []
for file in trace_files:
    with open(file, "r") as f:
        trace = [float(line.strip()) for line in f if line.strip()]
        traces.append(trace)

for i, t in enumerate(traces[:3]):
    print(f"Trace {i+1} (first 10 values):", t[:10])
```

```python
for trace in traces[:5]:
    plt.figure()
    plt.hist(trace, bins=100, edgecolor='black', alpha=0.7)
    plt.title("Data Distribution")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.show()
```

```python
%matplotlib ipympl
start_idx = 0
num_traces = 30
threshold = 0.4

for relative_t_idx, t in enumerate(traces[start_idx:start_idx + num_traces]):
    plt.figure()
    plt.title(f"Trace {start_idx + relative_t_idx}")
    plt.plot(t, alpha = 0.6)
    
    plt.axhline(threshold)
    plt.axhline(-threshold)

    plt.grid()
    plt.show()  
```
